```python
import os
import cv2
import tensorflow as tf
import numpy as np
from tqdm import tqdm
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.applications import MobileNetV2  # Import MobileNetV2 instead of VGG19
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Set the list of folders (classes)
folders = ['Potholes', 'Road Condition', 'Speedbumps']
# Load your dataset to Kaggle or update the path accordingly
data_dir = '/kaggle/input/indian-road-condition-dataset'

# Create empty lists to store images and labels
X = []
y = []

# Loop through each folder
for folder in folders:
    folder_path = os.path.join(data_dir, folder)

    for i in tqdm(os.listdir(folder_path)):
        img_path = os.path.join(folder_path, i)
        img = cv2.imread(img_path)

        # Check if the image was loaded successfully
        if img is not None:
            img = cv2.resize(img, (224, 224))  # Resize the image to MobileNetV2's input size (224x224) 
            X.append(img)
            y.append(folder)

# Convert X and y to NumPy arrays
X = np.array(X)
y = np.array(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Perform label encoding and one-hot encoding
le = preprocessing.LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)  # Use transform instead of fit_transform for the test set
y_train = tf.keras.utils.to_categorical(y_train, num_classes=len(folders))
y_test = tf.keras.utils.to_categorical(y_test, num_classes=len(folders))

# Create and compile the MobileNetV2-based model
mobilenet_model = Sequential()

# Load MobileNetV2 model
pretrained_model = MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')

for layer in pretrained_model.layers:
    layer.trainable = False

mobilenet_model.add(pretrained_model)
mobilenet_model.add(Flatten())
mobilenet_model.add(Dense(512, activation='relu'))
mobilenet_model.add(Dense(len(folders), activation='softmax'))  # Use len(folders) for the number of classes

# Compile the model
mobilenet_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
mobilenet_model.summary()

# Evaluate the model before training
y_pred_before_training = mobilenet_model.predict(X_test)
y_pred_before_training_classes = np.argmax(y_pred_before_training, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

accuracy_before_training = accuracy_score(y_test_classes, y_pred_before_training_classes)

print("Accuracy before training:", accuracy_before_training)

# Calculate confusion matrix before training
conf_matrix_before_training = confusion_matrix(y_test_classes, y_pred_before_training_classes)

# Plot confusion matrix before training
plt.figure(figsize=(6, 6))
plt.title('Confusion Matrix Before Training')
sns.heatmap(conf_matrix_before_training, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=folders, yticklabels=folders)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Train the model
epochs = 10 # You can adjust the number of epochs as needed
history = mobilenet_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=epochs)

# Evaluate the model after training
y_pred_after_training = mobilenet_model.predict(X_test)
y_pred_after_training_classes = np.argmax(y_pred_after_training, axis=1)

accuracy_after_training = accuracy_score(y_test_classes, y_pred_after_training_classes)
conf_matrix_after_training = confusion_matrix(y_test_classes, y_pred_after_training_classes)

print("\nAccuracy after training:", accuracy_after_training)
print("Confusion Matrix after training:")
print(conf_matrix_after_training)

# Plot training history (optional)
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0, 1])
plt.legend(loc='lower right')
plt.show()

# Plot confusion matrix after training
plt.figure(figsize=(6, 6))
plt.title('Confusion Matrix After Training')
sns.heatmap(conf_matrix_after_training, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=folders, yticklabels=folders)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show() 
```
